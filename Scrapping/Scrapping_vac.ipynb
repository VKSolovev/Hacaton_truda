Пишем функции, с помощью которых получаем зарплату, опыт, расписание, описание, ключевые навыки

```python
def salary(soup):
    s = soup.findAll('span', {'data-qa' : 'bloko-header-2', 'class' : 'bloko-header-2 bloko-header-2_lite'})
    if len(s) > 0:
        return s[0]
    else:
        trouble = True
```

```python
def experience(soup):
    s = soup.findAll('span', {'data-qa' : 'vacancy-experience'})
    if len(s) > 0:
        return s[0].text
    else:
        trouble = True
```

```python
def schedular(soup):
    s = soup.findAll('p', {'data-qa' : 'vacancy-view-employment-mode'})
    if len(s) > 0:
        return s[0].text
    else:
        trouble = True
```

```python
descriptions = soup.findAll('div', {'class' : 'g-user-content'})
```

```python
def description(soup):
    s = soup.findAll('div', {'class' : 'g-user-content'})
    if len(s) > 0:
        return s[0].text
    else:
        trouble = True
```

```python
def key_skills(soup): 
    return list(map(lambda x: x.text, soup.findAll('span', {'class' : 'bloko-tag__section bloko-tag__section_text', 'data-qa' : 'bloko-tag__text'})))
```

Получаем нужный soup по url

```python
def soup_url(url):
    headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36"}

    r = requests.get(url, headers=headers)
    return BeautifulSoup(r.text)
```

Читаем ссылки из файлов с вакансиями (что мы собирали в прошлом ноутбуке)

```python
def read_vacancy(name):
    f = open('url_vacancy\\' + name, 'r')
    res = []
    for line in f:
        res.append(line[:-1])
    return res
```

Получаем нужную информацию по url

```python
def get_info(url):
    
    soup = soup_url(url)
    sal = salary(soup)
    exp = experience(soup)
    if (len(soup_url(url).findAll('p', {'data-qa' : 'vacancy-view-employment-mode'})) > 0):
        sch = soup_url(url).findAll('p', {'data-qa' : 'vacancy-view-employment-mode'})[0].text
    else:
        trouble = True
        sch = ''
    des = description(soup)
    skill = key_skills(soup)
    return [sal, exp, sch, des, skill]
```

Создаем датафрейм из полученных данных

```python
def create_df():
    data = {'salary' : sal,
            'experience' : exp,
            'schedular' : sch,
           'description' : des,
           'key_skills' : skill,
           'url' : urls_df,
           'prof' : prof}
    return pd.DataFrame(data)
```

Теперь можем добавить данные, что мы получили к массивам, из которых потом сделаем датафрейм

```python
def add_file_data(file, dt_num):
    print(file)
    ammount_of_data = dt_num
    urls = read_vacancy(file)
    counter = 0
    for url in urls[0: min(len(urls), ammount_of_data)]:
        trouble = False
        counter += 1
        if counter % 1 == 0:
            print(str(counter) + '/' + str(min(len(urls), ammount_of_data)))
        arr = get_info(url)
        if not trouble:
            sal.append(arr[0])
            exp.append(arr[1])
            sch.append(arr[2])
            des.append(arr[3])
            skill.append(arr[4])
            urls_df.append(url)
            prof.append(file)
        else:
            print('Trouble with url: ' + url)
```

И наконец, запускаем все функции, что мы написали
